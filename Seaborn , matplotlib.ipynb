# Scatter Plot 

A scatterplot is a type of graph that displays the relationship between two numerical variables. It consists of points, each representing an observation, plotted on a Cartesian plane where one variable is on the x-axis and the other variable is on the y-axis. The pattern or trend of the points in the scatterplot helps to visually assess the correlation or relationship between the two variables.


There are 2 ways to draw a scatter plot in Seaborn --

1-relplot[Figure Level Function]

2-scatterplot[Axes Level Function]

# Boxplot

A boxplot, also known as a box-and-whisker plot, is a graphical representation of the distribution of a dataset based on five key summary statistics: minimum, first quartile (Q1), median (Q2), third quartile (Q3), and maximum. The box in the plot represents the interquartile range (IQR), which is the range between the first and third quartiles. The line inside the box represents the median value of the dataset. The "whiskers" extend from the box to the minimum and maximum values within a certain range, often defined as 1.5 times the IQR. Outliers, which are data points that fall outside this range, are plotted individually as points beyond the whiskers. Boxplots are useful for visualizing the spread, central tendency, and skewness of a dataset, as well as identifying potential outliers.






```python
#univariate

sns.boxplot(data['tip'])
```

```python
sns.catplot(x='day',y='total_bill',kind='box',data= data)
```

```python
sns.catplot(x='day',y='total_bill',kind='box',hue='sex',data= data)
```

So by using box plot using seaboarn library , we can see the distribution of the numerical data that is total_bill effectively.

# Heatmaps

A heatmap is a graphical representation of data where values are depicted using colors to visualize the magnitude of each value across different categories or dimensions. It typically consists of a grid where each cell's color intensity corresponds to the value of the data it represents. Heatmaps are commonly used to visualize relationships or patterns in large datasets, making it easier to identify trends or outliers visually.

```python
flight = sns.load_dataset('flights')

```

```python
flight.head()
```

```python
flight.pivot_table(index='year',columns= 'month',values='passengers',aggfunc='sum')
```

```python
x=flight.pivot_table(index='year',columns= 'month',values='passengers',aggfunc='sum')

sns.heatmap(x)
```

By seeing the heatmap we can conclude that from 1955 the customers for flight is increasing and the month of july and August has the maximum booking as compare to the the other month of the year.

```python
sns.heatmap(x,linewidths=0.6,annot=True,fmt='d')
```

```python
sns.heatmap(x,linewidths=0.6,annot=True,fmt='d',cmap='winter')
```

# Clustermap

Clustermap is same as heatmap but it will cluster the same columns

```python
sns.clustermap(x)
```

```python
sns.clustermap(x, z_score=0,annot= True,row_cluster=False,metric='correlation')
```

# Jointplots


A jointplot is a type of plot in statistical analysis that displays the relationship between two variables, usually by plotting them against each other and showing their correlation as well as the distribution of each variable separately. It often includes additional information such as regression lines or kernel density estimates to further explore the relationship between the variables.

We can use joinplots in 4 manners --

1-Scatterplots

2-Hexbinplots

3-Kdeplots

4-Regplots

```python
data.head()
```

```python
sns.jointplot(x='total_bill',y='tip',data=data)
```

Jointplot giving the scatterplot of total_bill and tip and indiviual histogram of each as well .

```python
#Hexplot

sns.jointplot(x='total_bill',y='tip',kind='hex',data=data)
```

```python
#2d Kdeplot

sns.jointplot(x='total_bill',y='tip',kind='kde',data=data)
```

```python
#regplot

sns.jointplot(x='total_bill',y='tip',kind='reg',data=data)
```

This is the best fit line  using the linear regression

# Pairplot


A pairplot is a type of plot in data visualization used to explore pairwise relationships between variables in a dataset. It displays a grid of subplots where each variable in the dataset is plotted against every other variable. Along the diagonal, instead of scatterplots, it shows the univariate distribution of each variable using histograms or kernel density estimates.

```python
iris = sns.load_dataset('iris')
```

```python
iris.head()
```

```python
sns.pairplot(iris)
```

```python
sns.pairplot(iris,hue='species')
```

# So , a very powerful tool to use is pairplot where all the columns are been corelated to each other and we can identify the correaltion on the basis of the scatterplot where the bar plot is for univariate data and  scatterplot is for bivariate data

# Distplots(Distribution Plots)

Distplot is use to do univariate data analysis of continous data

Displot combines 3 things in a graph ---

1-Histogram

2-Kernal Density Estimation (PDF)

3-Rugplot.

```python
titanic = sns.load_dataset('titanic')
```

```python
titanic.head()
```

```python
#For filling the missing values.


titanic['age'].fillna(titanic['age'].mean(),inplace = True)
```

# Plotting a Distplot

```python
sns.distplot(titanic['age'],rug= True)
```

# Only the histogram

```python
sns.distplot(titanic['age'],kde=False)
```

# Only the Kde

```python
sns.distplot(titanic['age'],hist= False)
```

# Plotting only Rugplot

```python
sns.distplot(titanic['age'],hist=False,kde =False , rug= True)
```

# How to use the Distplot

```python
sns.distplot(titanic[titanic['survived']==1]['age'])
```

# We are having a distplot of the age of the people who survived

```python
sns.distplot(titanic[titanic['survived']==1]['age'])
sns.distplot(titanic[titanic['survived']==0]['age'])
```

# Here is the graph where the people died resembles by red color we can clearly observe that  the probability of serviving is more for childrens than of adults.
